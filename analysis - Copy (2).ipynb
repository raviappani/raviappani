```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("cars_24original")
```

```python
def extract(x):
    if x<500000:
        return "Low"
    elif x<1000000:
        return "Medium"
    elif x<1500000:
        return "High"
    else:
        return "Very High"
    
```

```python
df["Price_level"]=df["Original_Price"].apply(extract)
```
