```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
#soup=BeautifulSoup(requests.get(link).text)
```

```python
price=[]
car=[]
payment=[]

for i in zipped:
    link=f"""https://www.cars24.com/buy-used-car?sort=bestmatch&serveWarrantyCount=true&gaId=1605453132.1699503151&storeCityId={i[0]}&pinId={i[1]}"""
    soup=BeautifulSoup(requests.get(link).text)
    for x in soup.find_all("div",attrs={"class":"_2z-Yu"}):
        cost=x.find("span",attrs={"class":"_18ToE"})
        name=x.find("h3",attrs={"class":"_2lmIw"})
        down=x.find("span",attrs={"class":"_2JX5N"})
        if cost is None:
            price.append(np.NaN)
        else:
            price.append(cost.text)
            
        if name is None:
            car.append(np.NaN)
        else:
            car.append(name.text)
            
        if down is None:
            payment.append(np.NaN)
        else:
            payment.append(down.text)
```

```python
df=pd.DataFrame({"car_name":car,"car_prices":price,"car_downpayment":payment})
```
