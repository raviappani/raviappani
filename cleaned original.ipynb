```python
city_id=[1692,4709,5732,132,5,3686,2378,2920,2130,6356,777,290,2,134,2423,136,769,6105,666,2713,2598,8184,1606,33,1605,1674]
pin_id=[380001,560001,60001,201001,122001,500001,400001,450001,302001,682001,700001,226001,110001,201301,411001,202124,\
       134101,641001,141001,440001,422001,800001,360001,124001,394101,390001]
zipped=list(zip(city_id,pin_id))
```

```python
print(zipped)
```

```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
price=[]
car=[]
payment=[]
ride_type=[]
fuel=[]
area=[]

for i in zipped:
    link=f"""https://www.cars24.com/buy-used-car?sort=bestmatch&serveWarrantyCount=true&gaId=1605453132.1699503151&storeCityId={i[0]}&pinId={i[1]}"""
    soup=BeautifulSoup(requests.get(link).text)
    for x in soup.find_all("div",attrs={"class":"_2z-Yu"}):
        cost=x.find("div",attrs={"class":"_18ToE"})
        name=x.find("h3",attrs={"class":"_2lmIw"})
        down=x.find("span",attrs={"class":"_2JX5N"})
        typ=x.find("ul",attrs={"class":"_1hOnS"})
        fue=x.find("ul",attrs={"class":"_13yb6"})
        are=x.find("span",attrs={"class":"_3DYbK"})

        if cost is None:
            price.append(np.NaN)
        else:
            price.append(cost.text)
            
        if name is None:
            car.append(np.NaN)
        else:
            car.append(name.text)
            
        if down is None:
            payment.append(np.NaN)
        else:
            payment.append(down.text)
            
        if typ is None:
            ride_type.append(np.NaN)
        else:
            ride_type.append(typ.text)
        if fue is None:
            fuel.append(np.NaN)
        else:
            fuel.append(fue.text)
        if are is None:
            area.append(np.NaN)
        else:
            area.append(are.text)
        
```

```python
len(payment)
```

```python
len(car)
```

```python
df=pd.DataFrame({"car_name":car,"car_prices":price,"car_downpayment":payment,"car_ride":ride_type,"car_fuel":fuel,\
                "car_area":area})
```

```python
df
```

```python
emi=[]

for i in zipped:
    link=f"""https://www.cars24.com/buy-used-car?sort=bestmatch&serveWarrantyCount=true&gaId=1605453132.1699503151&storeCityId={i[0]}&pinId={i[1]}"""
    soup=BeautifulSoup(requests.get(link).text)
    for x in soup.find_all("div",attrs={"class":"_2z-Yu"}):
        e=x.find("div",attrs={"class":"_1Em-A"})
        
        if e is None:
            emi.append(np.NaN)
        else:
            emi.append(e.text)
```
